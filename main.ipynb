# Diversification

## Diversifiable and Undiversifiable Risk

### 1. Diversifiable Risk (Unsystematic Risk)
- **Definition**: Risk specific to a company, industry, or sector. It can be reduced or eliminated through diversification (investing in a variety of assets).

- **Examples**:
  - **Company-Specific Events**: Poor management decisions, product recalls, lawsuits.
  - **Industry Events**: Regulatory changes affecting a specific sector, supply chain disruptions.

- **Characteristics**:
  - Not correlated with the overall market.
  - Can be minimized by holding a diversified portfolio (e.g., investing in multiple sectors or companies).
  - In a well-diversified portfolio, diversifiable risks are largely canceled out.

---

### 2. Undiversifiable Risk (Systematic Risk)
- **Definition**: Risk that affects the entire market or economy and cannot be eliminated through diversification. It is inherent to all investments.

- **Examples**:
  - **Economic Factors**: Inflation, interest rate changes, recessions.
  - **Political Risk**: Global conflicts, policy changes.
  - **Natural Disasters**: Pandemics or large-scale environmental events.

- **Characteristics**:
  - Correlated with the overall market.
  - Cannot be avoided regardless of how diversified the portfolio is.
  - Investors often seek higher returns to compensate for taking on systematic risk.

---

### How They Relate to Portfolio Theory
- **Portfolio Diversification**: By holding a variety of assets, investors can reduce diversifiable risk but not undiversifiable risk.
- **Beta**: In finance, the beta of an asset measures its sensitivity to systematic risk. A higher beta indicates higher exposure to market movements.

Understanding these distinctions helps investors manage their portfolios more effectively by diversifying to reduce unnecessary risks while accepting the systematic risks inherent to investing.





# Covariance and Correlation

### 1. Covariance
- **Definition**: Covariance measures how two random variables move together. It indicates whether the variables tend to increase or decrease at the same time.

- **Formula**:  
  $$ \text{Cov}(X, Y) = \frac{1}{n} \sum_{i=1}^{n} (X_i - \overline{X})(Y_i - \overline{Y}) $$

  Where:  
  - \( X_i \) and \( Y_i \) are individual data points.  
  - \( \overline{X} \) and \( \overline{Y} \) are the means of \( X \) and \( Y \).  
  - \( n \) is the number of data points.

- **Interpretation**:
  - **Positive Covariance**: Both variables tend to move in the same direction.
  - **Negative Covariance**: Variables move in opposite directions.
  - **Zero Covariance**: No relationship between the variables.

- **Limitation**: Covariance does not provide the strength or magnitude of the relationship, making it difficult to interpret its significance without additional context.

---

### 2. Correlation
- **Definition**: Correlation measures the strength and direction of the linear relationship between two variables. It is a standardized version of covariance.

- **Formula**:  
  $$ \text{Correlation}(X, Y) = \frac{\text{Cov}(X, Y)}{\sigma_X \cdot \sigma_Y} $$

  Where:  
  - \( \sigma_X \) and \( \sigma_Y \) are the standard deviations of \( X \) and \( Y \).

- **Interpretation**:
  - Values range from **-1** to **+1**:
    - **+1**: Perfect positive relationship.
    - **0**: No relationship.
    - **-1**: Perfect negative relationship.
  - Correlation is easier to interpret than covariance because it is dimensionless.

---

### **Key Differences**
- **Measurement**:
  - Covariance measures the direction of movement between two variables.
  - Correlation measures both the direction and strength of the relationship.
  
- **Range**:
  - Covariance has no fixed range.
  - Correlation always ranges from -1 to +1.

---

### **Example Use Case**
- In portfolio theory, **covariance** helps determine how assets move relative to each other, while **correlation** indicates how strongly the movements are related.
